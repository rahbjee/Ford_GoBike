<h1>Open-ended Questions</h1>
<h3>1.  What would you say about the status of the pilot to a stakeholder?</h2>

<p>In order to comment on the status of the pilot, I want to evaluate the number of trips that occur per month throughout the dataset.  Perhaps I will find a trend that I could report on.</p>
    <ul><li>Though we should note that there will be limited information from June 2017 since our earliest trips started on the 28th of the month</li></ul>

```python
CSV_data.index = CSV_data['start_time']
monthly_Data = CSV_data.groupby(by=CSV_data.index.month)
```

<p>Actually, on second thought, it would be nice to see the breakdown of monthly trips per user-type.  And plotting these trips on a histogram would make it easier to spot any trends of note</p>

```python
Sub_only.index = Sub_only['start_time'] #referring to subset created earlier in the notebook of only the subscriber trips
Sub_monthly_Data = Sub_only.groupby(by=Sub_only.index.month) #grouping by month  of start_time
Sub_count_series = Sub_monthly_Data.count()['bike_id']  #count the number of trips and 'bike_id' is an abritrary column to access single series from Groupby object (bit wonky)
Sub_count_series = Sub_count_series.rename("Subscriber_trips") #renaming to more understandable terms
Sub_count_series = Sub_count_series.rename(index={6: 'June', 7:'July',8:'August',9:'September',10:'October',11:'November',12:'December'})
Cust_only.index = Cust_only['start_time']  #doing same exact methods to customer only dataset
Cust_only_monthly_Data = Cust_only.groupby(by=Cust_only.index.month)
Cust_count_series = Cust_only_monthly_Data.count()['bike_id']
Cust_count_series = Cust_count_series.rename("Customer_trips")
Cust_count_series = Cust_count_series.rename(index={6: 'June', 7:'July',8:'August',9:'September',10:'October',11:'November',12:'December'})
total_count_df = pd.concat([Sub_count_series,Cust_count_series], axis=1)  #concating the two series into a DataFrame - that we can make a stacked barchart in next cell
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(12,4))  #Making subplots so that we can plot the monthly trips side by side
axe = ax.ravel()
all_ax = total_count_df.plot.bar(stacked=True,ax=axe[0])
all_ax.set_xlabel('Month of 2017')
all_ax.legend(loc='lower right')
all_ax.set_ylabel("# of trips")
Sub_count_series.plot.bar(color='blue',ax=axe[1]).set_xlabel('Month of 2017')
Cust_count_series.plot.bar(color='#ff7b00',ax=axe[2]).set_xlabel('Month of 2017')
fig.suptitle("Trips per month by user type")
```

<p>As mentioned, we can expect low numbers for June since there isn't much data from earlier in the month.  Still, a few notable observations include: </p>
    <ol><li>We see a promising upward trend in the number of trips for both subscribers and customers between July and October.</li>
        <li>Then we start to see a decline in the number of trips (both for subscribers and customers) from October to November and November to December.</li>
        <li>The decline in the number of trips is notably steeper for customers, whereas subscribers doesn't fall off as much and hold fairly steady near 70,000 trips per month towards the end of the year.</li>
    <li>During this phase in the pilot (from late June until the end of 2017), our subscribers take many more bike trips since our dataset contains almost 4 times more subscribers trips (409,230 vs. 110470) than customers.</li>

```python
Sub_only.shape #how many trips were taken by subscribers
```

```python
Cust_only.shape #how many trips were taken by customers only
```

<p>More questions arise from the above observations:  <ul><li>Is this decline due to seasonality of weather or tourism? <br/>(perhaps we can find data on the weather or tourism from this time - or we could conduct similar analysis on another year's data)</li><li>Or perhaps the number of trips correlates with the number of stations (we can create another variable for number of stations & conduct a similar analysis)?</li></p>

```python
numStationsPerMonth = monthly_Data['start_station_id'].nunique()
numStationsPerMonth = numStationsPerMonth.rename(index={6: 'June', 7:'July',8:'August',9:'September',10:'October',11:'November',12:'December'})
station_hist = numStationsPerMonth.plot.bar()
station_hist.set_xlabel('Month of 2017')
station_hist.set_ylabel("# of stations")
numStationsPerMonth
```

<p>By calculating & plotting these numbers in a histogram, we can see that the number of stations in use rises month over month throughout the whole date-range of this pilot dataset.<br/>  Though the initial rise in the number of stations available per month may account for some of the increase in the number of trips taken per month, we do not find a strong correlation because the number of stations in use continues to rise from October to December while the number of trips taken per month starts its decline.</p>

<h3>2. What users or people would you be interested in interviewing to learn more about their experience and why?</h3>
<p>I was highly intrigued by the range of the trip durations that we calculated in the 2nd warm-up question.  So I wanted to learn more about the distribution of trip duration.</p>

```python
average_duration_permonth = monthly_Data.describe(percentiles=[.1,.25,.5,.75,.9,.95,.99])['duration_min']
average_duration_permonth = average_duration_permonth.rename(index={6: 'June', 7:'July',8:'August',9:'September',10:'October',11:'November',12:'December'})
average_duration_permonth
```

```python
avg_duration_permon_subs = Sub_monthly_Data.describe(percentiles=[.1,.25,.5,.75,.9,.95,.99])['duration_min']
avg_duration_permon_subs = avg_duration_permon_subs.rename(index={6: 'June', 7:'July',8:'August',9:'September',10:'October',11:'November',12:'December'})
avg_duration_permon_subs
```

```python
avg_duration_permon_cust = Cust_only_monthly_Data.describe(percentiles=[.1,.25,.5,.75,.9,.95,.99])['duration_min']
avg_duration_permon_cust = avg_duration_permon_cust.rename(index={6: 'June', 7:'July',8:'August',9:'September',10:'October',11:'November',12:'December'})
avg_duration_permon_cust
```

<p>There is a clear separation between the user-types in the distribution of their trip durations.  <br/>Whereas 99% of subscriber trips fall under 45 minutes in duration, the top 10 percentile of customer trips (across all months) were greater than 60 minutes (or an hour)<br/>
I was curious about the costs of Ford's GoBike so I researched pricing and found information here: <a href="https://bikeshare.cc/guide/howtos/how-much-does-ford-gobike-cost/">How much does Ford GoBike Cost?</a><br/>Here I learned that there are: <ol><li>Single-Rides (what I will assume are considered Customers in this dataset), and </li><li>Access Pass or Monthly & Annual Membership plans (what I assume are considered Subscribers).</li></ol><br/>  I found it especially interesting given that Single-Rides are 2 dollars for the first 30 minutes and then 3 dollars per 15-minutes after that.  Therefore, a <i>two hours long single-ride</i> would cost: 2+3(6)= <b>20 dollars!</b></p>  I would be specifically interested in interviewing the Customers that fall within the top 5 percentile of trip duration since they for the most part (besides the month of November) rode for more than 120 minutes (2 hours).

<p>If 5 percent of the customers is too many people to interview, I certainly would like to interview the users who checked a bike our for the max duration for each month, which is consistently over 1000 minutes between June - December.  To calculate the price of a 1000-minute ride:  <br/>((1000min - 30min) / 15minutes )* 3 dollars)+ 2 dollars = <b>196 dollars!</b> for customers, or similarly <br/>((1000min - 45min) / 15minutes )* 3 dollars) = <b>191 dollars </b>for subscribers (on top of their associated subscription fee)
<br/>Overall, I'm curious about their riding experience and also whether they thought the costs were justified.</p>

```python
fig2, ax2 = plt.subplots(nrows=1, ncols=3, figsize=(12,4))
axe2 = ax2.ravel()
average_duration_permonth.drop(['count','max','99%','std','min'], axis=1, inplace=True)
all_users_duration_plot = average_duration_permonth.plot.bar(color=['black', 'red', 'orange', 'yellow', 'green', 'blue','purple'],ax=axe2[0])
all_users_duration_plot.set_xlabel('All users duration statistics')
all_users_duration_plot.set_ylabel('Trip Duration (in minutes)')
avg_duration_permon_subs.drop(['count','max','99%','std','min'], axis=1, inplace=True)
sub_dur_plot = avg_duration_permon_subs.plot.bar(color=['black', 'red', 'orange', 'yellow', 'green', 'blue','purple'],ax=axe2[1])
sub_dur_plot.get_legend().remove()
sub_dur_plot.set_xlabel('Subscribers trip duration')
avg_duration_permon_cust.drop(['count','max','99%','std','min'], axis=1, inplace=True)
cust_dur_plot = avg_duration_permon_cust.plot.bar(color=['black', 'red', 'orange', 'yellow', 'green', 'blue','purple'],ax=axe2[2])
cust_dur_plot.get_legend().remove()
cust_dur_plot.set_xlabel('Customers trip duration')
fig2.suptitle("Plotting percentiles of duration (in minutes) per user types")
```

<p>Although these charts are hard to read, if you note the scale of the y-axis for each graph you can tell that Customer trips were typically longer in duration.</p>

<h3>Also as someone who loves maps, I thought it would be noteworthy to plot the longest trips & the shortest trips (between different stations) in this dataset.</h3><p>For this I used the package geopandas to create a GeoDataFrame, and the package folium to plot it on a map of the Bay Area.</p>

```python
gdf = geopandas.GeoDataFrame(CSV_data, geometry=geopandas.points_from_xy(CSV_data.start_station_latitude, CSV_data.start_station_longitude))
gdf['geometryB'] = geopandas.points_from_xy(CSV_data.end_station_latitude,CSV_data.end_station_longitude)
gdf['same_station?'] = gdf['geometry'] == gdf['geometryB']
```

```python
gdf['distance'] = gdf.geometry.distance(gdf.geometryB)
```

```python
longest_trips_distance = gdf.sort_values(by='distance',ascending=False).iloc[0:10]  #creating another GeoDataframe of just the 10 longest trips in the dataset
longest_trips_distance
```

<p>Should note here that doing the similar calculation with shortest distance will often result in a 0 distance (meaning same station).  I could either:<ul><li>create another condition that it is the shortest but not yet zero, or</li><li>use the shortest duration trips (remember 61 seconds or just over 1 minute) that travelled from one station to different one.</li></ul>I chose the latter method because I was curious how these users were able to travel between stations so quickly!</p>

```python
min_duration_gdf = geopandas.GeoDataFrame(min_duration, geometry = geopandas.points_from_xy(min_duration.start_station_latitude, min_duration.start_station_longitude))
min_duration_gdf['geometryB'] = geopandas.points_from_xy(min_duration.end_station_latitude,min_duration.end_station_longitude)
min_duration_gdf['same_station?'] = min_duration_gdf['geometry'] == min_duration_gdf['geometryB']
oneMinDiff = min_duration_gdf[min_duration_gdf['same_station?'] == False]
oneMinDiff
```

<p>The above subset shows 13 trips that were 61 seconds in duration with different start / end stations.</p>
<p>Below I plot those 13 trips.  If a line is more opaque, then multiple trips occured between these two stations (for example: starting at S Van Ness at Market St & ending at Market St at 10th St.)

```python
mp2 = folium.Map(location=[37.75,-122.4], tiles='OpenStreetMap',zoom_start=12)
for ind,row in oneMinDiff.iterrows():
    tripGeo = [(row['start_station_latitude'],row['start_station_longitude']),(row['end_station_latitude'],row['end_station_longitude'])]
    folium.PolyLine(tripGeo, color='red', weight=10, opacity=0.3).add_to(mp2)
mp2
```

<p>I also ploted the 10 longest distance trips (calculated previously).  If a line is more opaque, then multiple trips occured between these two stations (for example: starting at Foothill Blvd at Fruitvale Ave & ending at Market St at 10th St.)

```python
mp3 = folium.Map(location=[37.67,-122.5], tiles='OpenStreetMap',zoom_start=9)
for ind,row in longest_trips_distance.iterrows():
    tripGeo = [(row['start_station_latitude'],row['start_station_longitude']),(row['end_station_latitude'],row['end_station_longitude'])]
    folium.PolyLine(tripGeo, color='green', weight=10, opacity=0.3).add_to(mp3)
mp3
```
